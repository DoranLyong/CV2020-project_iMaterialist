```python
import time
from tqdm.notebook import tqdm

for idx, img_ds in tqdm(enumerate(json_edit["images"]), desc="images"):
    json_fix["images"][idx]["id"] = idx + 41060
    
    for i, anno_ds in enumerate(json_edit["annotations"]):
        
        
        if json_edit["images"][idx]["id"] == json_edit["annotations"][i]["image_id"]:
            json_fix["annotations"][i]["image_id"] = idx + 41060
            
        
    
    
```
